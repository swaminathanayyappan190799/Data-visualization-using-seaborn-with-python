```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.boxplot(x='sex',y='tip',data=tips)
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
import numpy as np
sns.barplot(x='sex',y='tip',data=tips,estimator=np.mean)
```

```python
sns.violinplot(x='sex',y='tip',data=tips,hue='sex',palette='coolwarm',split=True)
```

```python
sns.swarmplot(x='sex',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.scatterplot(x='sex',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.stripplot(x='sex',y='tip',data=tips,hue='sex',palette='coolwarm')
```
